```python
# 51개의 선택지를 '한식','중식','일식','양식','아시안식'으로 나눔
한식 = ['간장게장', '골뱅이무침', '김치찌개', '너구리', '닭도리탕', '닭백숙', '닭볶음탕','대구맑은탕', '대구매운탕', '대구지리',
     '돼지갈비', '된장찌개', '떡볶이', '매운칼국수', '물냉면', '바지락칼국수','보쌈', '볶음밥','비빔냉면','비빔면', '비빔밥',
      '삼결살', '삼겹살', '순두부찌개','신라면', '양념게장', '양념치킨', '양평해장국', '어묵','족발', '짜파게티', '청국장찌개',
     '콩나물국밥','파전', '후라이드']
중식 = ['군만두', '물만두','양장피', '자장면','짬뽕','탕수육(부먹)', '탕수육(찍먹)','팔보채']
일식 = ['돈가스', '일본라멘', '일본식우동']
양식 = ['봉골레','스테이크','카르보나라','해물토마토파스타']
아시안식 = ['베트남쌀국수']
```

```python
# 자카드 유사도
def similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return round(float(len(s1.intersection(s2)) / len(s1.union(s2))),2)
```

```python
# 교집합
def intersection(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return s1.intersection(s2)
```

```python
# 자카드 유사도 기반의 유사도
print("1번 사람과 2번 사람의 유사도 : ", round(similarity(food_22.iloc[0][1],food_22.iloc[1][1]),2))
```

```python
# 교집합
print("1번 사람과 2번 사람의 공통 선택지 : ", intersection(food_22.iloc[0][1],food_22.iloc[1][1]))
```

### 1번 사람과 2번 사람의 선택지를 참고하여 '한식','중식','일식','양식','아시안식'을 기반으로 음식 카테고리 추천

```python
test = intersection(food_22.iloc[0][1],food_22.iloc[1][1])
```

```python
def choose(same):
    
    print("공통적으로 선택한 음식과 그 수")
    print("")
    
    kr_food = list(set(한식) & set(same))          ## 공통 선택한 한식의 종류
    kr_food_num = len(list(set(한식) & set(same))) ## 공통 선택한 한식의 수
    kr_s = round((kr_food_num) / len(set(food_list)),2) ## 공통된 한식을 선택한 확률 ex) 선택한 한식 / 전체 한식
    print('한식 : ',kr_food, kr_food_num,"개")
    print('한식 유사도 : ', kr_s)
    
    print("")
    
    ch_food = list(set(중식) & set(same))          ## 공통 선택한 중식의 종류
    ch_food_num = len(list(set(중식) & set(same))) ## 공통 선택한 중식의 수
    ch_s = round((ch_food_num) / len(food_list),2) ## 공통된 중식을 선택한 확률
    print('중식 : ',ch_food, ch_food_num,"개")
    print('중식 유사도: ', ch_s)
    
    print("")
    
    jp_food = list(set(일식) & set(same))          ## 공통 선택한 일식의 종류
    jp_food_num = len(list(set(일식) & set(same))) ## 공통 선택한 일식의 수
    jp_s = round((jp_food_num) / len(food_list),2) ## 공통된 일식을 선택한 확률
    print('일식 : ',jp_food, jp_food_num,"개")
    print('일식 유사도: ', jp_s)
    
    print("")
    
    ws_food = list(set(양식) & set(same))          ## 공통 선택한 양식의 종류
    ws_food_num = len(list(set(양식) & set(same))) ## 공통 선택한 양식의 수
    ws_s = round((ws_food_num) / len(food_list),2) ## 공통된 양식을 선택한 확률
    print('양식 : ',ws_food, ws_food_num,"개")
    print('양식 유사도: ', ws_s)
    
    print("")
    
    as_food = list(set(아시안식) & set(same))          ## 공통 선택한 아시안식
    as_food_num = len(list(set(아시안식) & set(same))) ## 공통 선택한 아시안식의 수
    as_s = round((as_food_num) / len(food_list),2) ## 공통된 아시안식을 선택한 확률
    print('아시안식 : ',as_food, as_food_num,"개")
    print('아시안 유사도: ', as_s)
    
    print("")
    
    # food_cat에 전체 확률 값을 리스트 형식으로 구성
    food_cat = [kr_s, ch_s, jp_s, ws_s, as_s]
    
    # food_cat에서 가장 큰 확률값의 index를 출력
    max_cat = food_cat.index(max(food_cat))
        
    if max_cat == 0:
        print("한식을 추천합니다.")

    elif max_cat == 1:
        print("중식을 추천합니다.")

    elif max_cat == 2:
        print("일식을 추천합니다.")

    elif max_cat == 3:
        print("양식을 추천합니다.")

    elif max_cat == 4:
        print("아시안식을 추천합니다.")
```

```python
choose(test)
```
