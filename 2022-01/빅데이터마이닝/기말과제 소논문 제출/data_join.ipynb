## 데이터 합치기

```python
merge_m = pd.merge(train, meal)
merge_m
```

```python
merge_c = pd.merge(train, center)
merge_c
```

```python
merge_mc = pd.merge(merge_m, merge_c)
merge_mc
```

```python
train_id = pd.merge(train, merge_mc, on='id', how='left')
train_id
```
