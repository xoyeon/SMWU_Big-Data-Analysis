### [실습] 프린트 실습 (2분)
* a에 3432*1234 의 결과 값을 넣고 print(f"3432곱하기 1234값은 ....)으로 결과 값을 출력하시오

```python
import time

def timers():
  in_min = input("시간을 입력하세요.(분):")
  min = int(in_min)
  sec = min*60
  print(sec)

  #while은 반복문으로 sec가 0이 되면 반복을 멈춰라
  while (sec != 0 ):
      sec = sec-1
      time.sleep(1)
      print(sec)

timers()
```

# 11.다중 조건문 (if, elif, else)

if 조건1:

  <조건1일때 실행>

elif 조건2:

  <조건2일때 실행>

else:

  <이외일때 실행>

```python
a = input("첫번째 숫자를 입력하세요: ")
a = int(a)

b = input("두번째 숫자를 입력하세요: ")
b = int(b)


if(a<b):
  print(f"처음입력수 {a}보다 두번째 입력수 {b}가 큽니다.")
elif(a>b):
  print(f"처음입력수 {a}보다 두번째 입력수 {b}가 작습니다.")
else:
  print(f"두 수가 {a}동일합니다.")
```

### [실습] 다중 조건문에 대해 실습합니다. (10분)
* aa에 리스트로 "구본무", "김범수", "구현모", "이건희" 를 저장합니다.
* bb에 리스트로 "박혜일", "전미도", "조정석", "김혜수" 를 저장합니다.
* cc에 사람이름을 입력합니다. (키보드로 입력받습니다.) 
* 입력받은 사람이름이 aa에 있으면 "우리나라 CEO입니다" 를 bb에 있으면"연애인입니다." 그렇지 않으면" "CEO 도 연예인도가아닙니다"를 출력합니다.
* 아래 코드를 활용하세요

```python
aa = ["구본무", "김범수", "구현모", "이건희" ]
bb = ["박혜일", "전미도", "조정석", "김혜수" ]
cc = input("이름을 입력하세요 : ")

if cc in aa:
  print(f"{cc}는 CEO입니다.")
elif cc in bb:
  print(f"{cc}는 연예인입니다.")
else:
  print(f"{cc}는 연예인도 CEO도 아닙니다.")
```

# **[과제1] 지금까지 배운 것을 종합적으로 실습해 봅니다.(15분)**

* 지금까지 배운 것으로 다음을 실습해 보시기 바랍니다.

* input()을 이용해서 숫자 입력을 받고, 변수 a에 이를 넣어봅시다. 
* 입력 받은 수가 만약 실수(예: 1.1, 0.5등) 이면 이를 반올림하여 정수로 변환합니다.
* if-else문을 이용해서 만약 입력받은 수가 홀수면 "(입력받은 수)는 홀수입니다." 짝수면 "(입력받은 수) 짝수입니다."를 출력해봅니다.

* 1) 수업날짜_본인이름.ipynb (예: 0302_김OO.ipynb)

* 2) 스노우에 제출바랍니다.

* 3) 기한: 매주 해당 수업 수강 이후 24h 이내

```python
a = input("숫자를 입력하시오 : ")
a = float(a)
a = round(a)
a = int(a)

print(a)

if a % 2 == 1:
  print(f"{a}는 홀수입니다.")
else:
  print(f"{a}는 짝수입니다.")
```
