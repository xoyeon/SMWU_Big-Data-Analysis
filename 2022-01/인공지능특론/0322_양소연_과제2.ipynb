# 3.반복문(While)
* while 조건: (조건이 참일때만 진행함)
* break를 사용하여, 특정 조건일때 반복을 빠져나오게함
* break를 이용하여, 리스트안에서, 수박을 찾고, 찾으면 빠져나옴


```python
fruit = ['사과', '귤', '포도', '수박', '복숭아']
```

```python
print("수박" in fruit)
```

```python
i=0
while True:
  print(fruit[i])
  if(fruit[i] == "수박"):
    break
  i=i+1
print("수박이 나왔네요")
```

```python
# 숫자를 입력하다가 input 리스트에 계속 추가하다가 "0"이 입력되면 stop한다.

input_list = []

while True:
  a = input("입력하시오 : ")
  if a == "0":
    break
  input_list.append(a)
```

```python
print(input_list)
```

# **[과제2] While를 활용한 데이터 입력 후 평균값구하기**
* -1이 입력 될때 까지 정수를 입력받아 input_value에 리스트로 저장한다.
* 입력 받은 자료를 모두 더한후 평균 값을 구하세요

* 1) 수업날짜_본인이름.ipynb (예: 0302_김OO.ipynb)

* 2) 스노우에 과제제출

* 3) 기한: 매주 해당 수업 수강 이후 24h 이내

```python
input_value = []

while True:
  a = input("정수를 입력하시오 : ")
  if a == "-1":
    break
  input_value.append(a)

# 리스트 문자열을 int로 바꾸기
input_int = list(map(int, input_value))
print(input_int)
```

```python
print("리스트의 합 : ", sum(input_int))
print("리스트의 평균 : ", sum(input_int)/len(input_int))
```
