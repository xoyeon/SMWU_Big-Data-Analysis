# **[2] 데이터 기반으로 학습데이터/테스트데이터로 구분하기**

* 아래 코드 그대로 활용한 후

#### *(필수코드)*

```python
# 필수코드로, 이미지데이터를 변환하고 학습, 검증데이터로 분리하는 코드임
# 변경사항은 분류폴더가 있는 상위폴더명 (face_dir = "/content/cat_dog"), 카테고리명(categories = ["cats","dogs"])
# 그리고 np.save("2obj.npy", xy)도 원하는 파일로 변경필요


from PIL import Image
import os, glob
import numpy as np
from sklearn.model_selection import train_test_split


face_dir = '/content/drive/MyDrive/Colab Notebooks/밥/'
categories = ["백미","현미", "흑미"]
nb_classes = len(categories)
# 이미지 크기 지정 
image_w = 64 
image_h = 64
pixels = image_w * image_h * 3

# 이미지 데이터 읽어 들이기 
X = []
Y = []

for idx, cat in enumerate(categories):
    # 레이블 지정으로 모두 0으로 지정하고 현재의 카테고리는 1로 지정하여 원핫인코딩하는 것임 (원핫인코깅해도됨)
    label = [0 for i in range(nb_classes)]
    label[idx] = 1
    # 이미지 
    image_dir = face_dir + "/" + cat
    files = glob.glob(image_dir+"/*.jpg")
    for i, f in enumerate(files):
        img = Image.open(f) 
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)      # numpy 배열로 변환
        X.append(data)
        Y.append(label)
        if i % 10 == 0:
            print(i, "\n", data, label)
X = np.array(X)
Y = np.array(Y)

# 학습 전용 데이터와 테스트 전용 데이터 구분 , 0.25가 디폴트임
X_train, X_test, y_train, y_test = \
    train_test_split(X, Y)
xy = (X_train, X_test, y_train, y_test)

print('>>> data 저장중 ...')
np.save("2obj.npy", xy)
print("ok,", len(Y))
```

# **[3] 모델링 및 학습하기**

#### *(필수코드)*

```python
# 본 코드는 cnn 적용하여 분류문제로 모델링 한 내용임
# 변경할 부분은 categories = ["cats","dogs"] 코드와 아래 레이어 필요시 추가, h_file = "4obj-model.h" 저장할 모델 파일명,
# model.fit(X_train, y_train, batch_size=32, epochs=30)에서 에폭수와 배치수 정도만 변하면 됨

from tensorflow import keras
from keras.models import Sequential
from keras.layers import MaxPooling2D
from keras.layers import Conv2D
from keras.layers import Activation, Dropout, Flatten, Dense
import numpy as np
import os


# 카테고리 지정하기
categories = ["백미","현미", "흑미"]
nb_classes = len(categories)
# 이미지 크기 지정하기
image_w = 64
image_h = 64
# 데이터 열기 
#X_train, X_test, y_train, y_test = np.load("2obj.npy")
# 데이터 정규화하기(0~1사이로)
X_train = X_train.astype("float") / 256
X_test  = X_test.astype("float")  / 256
print('X_train shape:', X_train.shape)

# 모델 구조 정의 
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=X_train.shape[1:], padding='same'))
model.add(Activation('relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))

model.add(Conv2D(128, (3, 3), padding='same')) # 64를 128로 변경, 패딩추가
model.add(Activation('relu')) # 할성함수 추가

model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))


# 전결합층
model.add(Flatten())    # 벡터형태로 reshape
model.add(Dense(512))   # 출력
model.add(Dense(128))   # 추가
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(nb_classes))
model.add(Activation('softmax'))
# 모델 구축하기
model.compile(loss='categorical_crossentropy',   # 최적화 함수 지정
    optimizer='rmsprop',
    metrics=['accuracy'])
# 모델 확인
#print(model.summary())

# 학습 완료된 모델 저장
h_file = "/content/drive/MyDrive/Colab Notebooks/4obj-model.h"
if os.path.exists(h_file):
    # 기존에 학습된 모델 불러들이기
     model = keras.models.load_model(h_file)
else:
    # 학습한 모델이 없으면 파일로 저장
    model.fit(X_train, y_train, batch_size=32, epochs=30)
    model.save(h_file)
```

### *(필수코드)*

```python
# 테스트 데이터 셋을 통해 모델 평가함
# 변경은 여기서는 불필요함

score = model.evaluate(X_test, y_test)
print('loss=', score[0])        # loss
print('accuracy=', score[1])    # acc
```
