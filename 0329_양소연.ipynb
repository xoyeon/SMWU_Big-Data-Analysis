```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [1] 다음 파일을 열어 주세요. (Employee Data.csv)
data = pd.read_csv("/content/drive/MyDrive/2022-01/인공지능특론/4. Employee Data.csv", sep=',')
data.head()
```

```python
# [2] Y는 Attrition(퇴사여부) 이고 나머지는 X 입니다.
data['Attrition'].value_counts().plot(kind = 'bar')
```

```python
# [3] Gender에 따른 퇴사(Attrition) 여부 갯수를 구해주세요
female_yes = data[(data['Gender'] == 'Female') & (data['Attrition'] == 'Yes')]
female_no = data[(data['Gender'] == 'Female') & (data['Attrition'] == 'No')]
male_yes = data[(data['Gender'] == 'Male') & (data['Attrition'] == 'Yes')]
male_no = data[(data['Gender'] == 'Male') & (data['Attrition'] == 'No')]

print("퇴사한 여성 : ", female_yes.shape[0], "명")
print("퇴사하지 않은 여성 : ", female_no.shape[0], "명")
print("퇴사한 남성 : ", male_yes.shape[0], "명")
print("퇴사하지 않은 남성 : ", male_no.shape[0], "명")
```

```python
# [4] Attrition 즉, 퇴사가 Yes인 행들만 도출 바랍니다.
data.loc[(data['Attrition'] == 'Yes')]
```

```python
# [5] 성별이 남성(Male)이면서 불만(Low)하고, 홈오피스가 먼(Far) 사람들의 퇴사건수를 구해주세요
print(data[(data['Gender'] == 'Male') & (data['Satisfaction'] == 'Low') & (data['Home-Office'] == 'Far')].shape[0],"건")
```
